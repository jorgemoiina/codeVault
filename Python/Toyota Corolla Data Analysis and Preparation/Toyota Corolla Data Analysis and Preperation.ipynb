correlation matrix + matrix plot below

```python
numeric_columns = toyotaDf.select_dtypes(include=['int64', 'float64']).columns
numeric_toyotaDf = toyotaDf[numeric_columns]
```

```python
correlation_matrix = numeric_toyotaDf.corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('correlation matrix')
plt.show()
```
