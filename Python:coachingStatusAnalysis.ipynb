```python
import pandas as pd
import pyperclip
```

```python
link  = r'C:\Users\imolinjo\Downloads\priorday.csv' # this can vary from "priorday", "priorweek", or any other specified timeframe
DF = pd.read_csv(link)
```

```python
print(DF.columns)
```

```python
print(DF[['Supervisor Name', 'Coaching Status']])
```

```python
groupcounts = DF.groupby(['Supervisor Name', 'Coaching Status']).size()
print(groupcounts)
```

```python
groupcounts = DF.groupby(['Supervisor Name', 'Coaching Status']).size().reset_index(name='Count')
output = groupcounts.to_csv(sep='\t', index=False)  # convert DF into a string
pyperclip.copy(output) # i opt to have this posted to my clipboard due to me manually "pasting" it into an existing Excel sheet
print(output)  
```

```python
groupcountsCancel = DF.groupby(['Supervisor Name', 'Coaching Status', 'Closed Reason']).size()
print(groupcountsCancel)
```

```python
filtered_groupcounts = groupcountsCancel[~groupcountsCancel.index.get_level_values('Closed Reason').isin(['AUTO_CANCELLED_ASSOCIATE_NOT_PUNCHED_IN', 'AUTO_CANCELLED_TOO_MANY_COMPLETED_COACHING', 'AUTO_CANCELLED_MISSING_PUNCH_DATA'])]
print(filtered_groupcounts)
```
