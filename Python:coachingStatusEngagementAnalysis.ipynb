```python
import pandas as pd
import math
import pyperclip
```

```python
link  = r'C:\Users\imolinjo\Downloads\ohdire.csv'
DF = pd.read_csv(link)
```

```python
print(DF.head())
```

```python
DFcolumns = (DF.columns)
print(DFcolumns)
```

```python
print(DF[['Department', 'Category']])
```

```python
groupcounts = DF.groupby(['Department', 'Category']).size()
print(groupcounts)
```

```python
selectedcategories = ['PICK_SHORT', 'PICK_REJECT', 'RECEIVE_ERROR_INDICATOR', 'RECEIVE_QUANTITY_ERROR', 
                     'SHIP_LOAD_QUALITY', 'STOW_BIN_COLLISION', 'STOW_SCAN_OUT_OF_SEQUENCE',
                      'STOW_SIPS_OVER_AND_SHORT', 'PICK_FALSE_SHORT', 'PACK_SINGLE_SELF_SELECTED_SIOC']
selectCounts = groupcounts[groupcounts.index.get_level_values('Category').isin(selectedcategories)]
print(selectCounts)
```

```python
selectCounts = groupcounts[groupcounts.index.get_level_values('Category').isin(selectedcategories)].reset_index()
output = selectCounts.to_csv(sep='\t', index=False)
pyperclip.copy(output) # i opt to have this posted to my clipboard due to me manually "pasting" it into an existing Excel sheet
print(output) 
```
